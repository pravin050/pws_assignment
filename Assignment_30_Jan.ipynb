#### Q4. Check how many timnes a given number can be divided by 3 before it is less than or equal to 10

```python
num = int(input("Enter anumber:"))

count = 0
while num > 10:
    num = num / 3
    count += 1
    
print("Given number can be divided by 3 before it become less than or equal to 10 is", count ,"times.")
```

#### A while loop in Python is used to repeatedly execute a block of code as long as a given condition is true. This type of loop is called a "conditional loop" because the number of iterations is determined by a condition that is tested before each iteration.

below is exmaple

### Use nested while loop to print 3 different pattern

```python
# Pattern 1: Triangle
rows = 5
print ('\n\n')
for i in range(rows):
    for j in range(i + 1):
        print("*", end="")
    print("")
print ("\n\n\n")

# Pattern 2: Diamond
rows = 5

for i in range(1, rows + 1):
    spaces = rows - i
    for j in range(spaces):
        print(end=" ")
    for j in range(2 * i - 1):
        print("*", end="")
    print("")

for i in range(rows - 1, 0, -1):
    spaces = rows - i
    for j in range(spaces):
        print(end=" ")
    for j in range(2 * i - 1):
        print("*", end="")
    print("")
print ("\n\n\n")

    
# Pattern 4: Pyramid
rows = 5

for i in range(1, rows + 1):
    spaces = rows - i
    for j in range(spaces):
        print(end=" ")
    for j in range(2 * i - 1):
        print("*", end="")
    print("")
print ("\n\n\n")

```
